```python
#문제1
from pyformlang.finite_automaton import DeterministicFiniteAutomaton
from pyformlang.finite_automaton import State
from pyformlang.finite_automaton import Symbol
from itertools import product

# Declaration of the DFA
dfa = DeterministicFiniteAutomaton()

# Creation of the states
state0 = State(0)
state1 = State(1)
state2 = State(2)
# Creation of the symbols
symb_0 = Symbol("0")
symb_1 = Symbol("1")

# Add a start state
dfa.add_start_state(state0)

# Add two final states
dfa.add_final_state(state1)

# Create transitions
dfa.add_transition(state0, symb_0, state0)
dfa.add_transition(state0, symb_1, state1)
dfa.add_transition(state1, symb_1, state2)
dfa.add_transition(state2, symb_0, state2)
dfa.add_transition(state2, symb_1, state1)
dfa.add_transition(state1, symb_0, state0)

for i in range(4,11):
    for x in product(["0","1"], repeat=i):
        y=list(x)
        print(y)
        print(dfa.accepts(y)) 
            
            
```

```python
#문제3
from pyformlang.finite_automaton import NondeterministicFiniteAutomaton
from pyformlang.finite_automaton import State
from pyformlang.finite_automaton import Symbol

# Definition of the NFA
nfa = NondeterministicFiniteAutomaton()

# Declare the states
state0 = State(0)
state1 = State(1)
state2 = State(2)

# Declare the symbols
symb_0 = Symbol("0")
symb_1 = Symbol("1")
symb_λ = Symbol("λ")

# Add a start state
nfa.add_start_state(state0)

# Add a final state
nfa.add_final_state(state1)

# Add the transitions
nfa.add_transition(state0, symb_0, state1)
nfa.add_transition(state0, symb_1, state1)
nfa.add_transition(state1, symb_0, state0)
nfa.add_transition(state1, symb_1, state1)
nfa.add_transition(state1, symb_0, state2)
nfa.add_transition(state1, symb_λ, state2)
nfa.add_transition(state2, symb_1, state1)

x={'00', '010011', '10010', '000', '0000'}
for i in x:
    
    print(i,nfa.accepts(i))
```
